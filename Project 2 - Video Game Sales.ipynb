This dataset contains a list of video games with sales greater than 100,000 copies. It was generated by a scrape of vgchartz.com.

Fields include

Rank - Ranking of overall sales

Name - The games name

Platform - Platform of the games release (i.e. PC,PS4, etc.)

Year - Year of the game's release

Genre - Genre of the game

Publisher - Publisher of the game

NA_Sales - Sales in North America (in millions)

EU_Sales - Sales in Europe (in millions)

JP_Sales - Sales in Japan (in millions)

Other_Sales - Sales in the rest of the world (in millions)

Global_Sales - Total worldwide sales.


```python
#Library required and Dataset is imported

#We need matplot to plot the data in for Iris dataset
import matplotlib.pyplot as plt

# Let's import python first as it an important package for data processing
import pandas as pd

# This method produces a lot of warning, so let's import the package and ignore those warnings 
import warnings 
warnings.filterwarnings("ignore")

# This handles  the error of jyputer not plotting
%matplotlib inline
```

```python
# lets load the iris data set using pandas
vgsales = pd.read_csv("C:\\Users\\pnkum\\Desktop\\Project\\Data-Visualization\\vgsales.csv") 

# Lets us see the first 10 entries of the file.
vgsales.head(15)
```

Before we continue to plot different graphs on this dataset, It's important to check if there are any NA or NaN. If we find any we need to remove.

```python
vgsales.isnull().values.any()
```

```python
vgsales = vgsales.dropna()
vgsales.isnull().values.any()
```

Now That we have removed all the NA we can now plot the data.
## We will be plotting barplot of Genre and number of sales

### What is a barplot?
A bar chart or bar graph is a chart or graph that presents categorical data with rectangular bars with heights or lengths proportional to the values that they represent. The bars can be plotted vertically or horizontally.

soruce : https://en.wikipedia.org/wiki/Bar_chart

```python
fig,ax = plt.subplots(figsize=(10,10))
vgsales['Genre'].value_counts(sort=False).plot(kind='bar',color = 'red')
plt.title('Genre Distribution',fontsize=15)
plt.xlabel('Genre',fontsize=15)
plt.ylabel('Number of sales',fontsize=15)
```

As defined above the barplot is alsi vaviable horizontally. we will try to explore the horizontal bar plot

```python
fig,ax = plt.subplots(figsize=(10,10))
vgsales['Genre'].value_counts(sort=False).plot(kind='barh',color = 'red')
plt.title('Genre Distribution',fontsize=15)
plt.xlabel('Genre',fontsize=15)
plt.ylabel('Number of sales',fontsize=15)
```

I am also interested  to check top 5 years where highest sale has happened, so we plot a barplot and sort the data

## We will be plotting barplot of Year and number of sales

```python
fig,ax = plt.subplots(figsize=(13,13))
vgsales['Year'].value_counts(sort=True).plot(kind='bar',color = "red")
plt.title('Year Distribution',fontsize=15)
plt.xlabel('Year',fontsize=15)
plt.ylabel('Number of sales',fontsize=15)
```

We see that year 2009, 2008, 2010 2007 and 2011 had the highest number of sales

## We will be plotting pie chart of Platform and number of sales

### what is pie chart?

A pie chart (or a circle chart) is a circular statistical graphic which is divided into slices to illustrate numerical proportion. In a pie chart, the arc length of each slice (and consequently its central angle and area), is proportional to the quantity it represents. While it is named for its resemblance to a pie which has been sliced, there are variations on the way it can be presented.

Source : https://en.wikipedia.org/wiki/Pie_chart 

```python
fig,ax = plt.subplots(figsize=(15,15))
vgsales['Platform'].value_counts(sort=False).plot(kind='pie')
plt.title('Platform Distribution',fontsize=15)
plt.xlabel('Platform',fontsize=15)
plt.ylabel('Number of sales',fontsize=15)
```

## We will be plotting pie chart of Genre and number of sales

```python
fig,ax = plt.subplots(figsize=(15,15))
vgsales['Genre'].value_counts(sort=False).plot(kind='pie')
plt.title('Platform Distribution',fontsize=15)
plt.xlabel('Platform',fontsize=15)
plt.ylabel('Number of sales',fontsize=15)
```

## We will be plotting boxplot of Genre and number of sales

### what is boxplot?

The box plot (a.k.a. box and whisker diagram) is a standardized way of displaying the distribution of data based on the five number summary: minimum, first quartile, median, third quartile, and maximum.

Source : http://www.physics.csbsju.edu/stats/box2.html

```python
fig,ax = plt.subplots(figsize=(10,10))
vgsales['Genre'].value_counts(sort=False).plot(kind='Box', color = 'red')
plt.title('Genre Distribution',fontsize=15)
plt.xlabel('Genre',fontsize=15)
plt.ylabel('Number of sales',fontsize=15)
```

We see one of the genre is an outliner

## We will be plotting area chart of Platform and number of sales
### what is an area chart?
An area chart or area graph displays graphically quantitative data. It is based on the line chart. The area between axis and line are commonly emphasized with colors, textures and hatchings. Commonly one compares with an area chart two or more quantities.

Source : https://en.wikipedia.org/wiki/Area_chart

```python
fig,ax = plt.subplots(figsize=(12,12))
vgsales['Platform'].value_counts(sort=False).plot(kind='area', color = 'red')
plt.title('Platform Distribution',fontsize=15)
plt.xlabel('Platform',fontsize=15)
plt.ylabel('Number of sales',fontsize=15)
```

## We will be plotting kde of Platform and number of sales

```python
fig,ax = plt.subplots(figsize=(10,10))
vgsales['Platform'].value_counts(sort=False).plot(kind='kde',color = 'red')
plt.title('Platform Distribution',fontsize=15)
plt.xlabel('Platform',fontsize=15)
plt.ylabel('Number of sales',fontsize=15)
```
